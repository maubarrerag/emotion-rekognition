```python
import boto3
import os
```

```python
path = os.getcwd()+'/lfw/'
print(path)
```

```python
files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.jpg' in file:
            files.append(os.path.join(r, file))
```

```python
print(len(files))
```

```python
bucket = 'face-recognition-mbg'
s3 = boto3.resource('s3')

def print_progress(progress):
    print(progress)
```

```python
for file in files:
    (base, f) = os.path.split(file)
    print("Uploading:"+f)
    s3.meta.client.upload_file(Filename=file, Bucket=bucket, Key=f, Callback=print_progress)
```
